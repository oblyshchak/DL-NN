```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_1 = cv2.imread('chess_1.png')
img_2 = cv2.imread('chess_2.png')
```

```python
plt.figure(figsize=(12,8))
pts_dst = np.array([[410, 75],[400, 225],[210, 120],[210, 240]])
plt.imshow(img_1)
plt.show()
```

```python
plt.figure(figsize=(12,8))
pts_src = np.array([[215, 120], [410, 150], [275, 340],[390, 350]])
plt.imshow(img_2)
plt.show()
```

```python
h, status = cv2.findHomography(pts_src, pts_dst)
```

```python
im_out = cv2.warpPerspective(img_2, h, (img_1.shape[1],img_1.shape[0]))
```

```python
plt.figure(figsize=(12,8))
plt.imshow(im_out)
plt.show()
```

```python
rows, cols  = im_out.shape[:2]
```

```python
angle = -260
```

```python
M = cv2.getRotationMatrix2D((cols/2, rows/2), angle, 1)
```

```python
rotated_img = cv2.warpAffine(im_out, M, (cols, rows))
```

```python
plt.figure(figsize=(12,8))
plt.imshow(rotated_img)
plt.show()
```
