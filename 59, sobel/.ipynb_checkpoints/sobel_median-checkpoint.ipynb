```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('hand.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
median = cv2.medianBlur(img, 15)
```

```python
gX = cv2.Sobel(img, ddepth=cv2.CV_32F, dx=1, dy=0, ksize=3)
gY = cv2.Sobel(img, ddepth=cv2.CV_32F, dx=0, dy=1, ksize=3)
gX = cv2.convertScaleAbs(gX)
gY = cv2.convertScaleAbs(gY)
```

```python
img_sobel = cv2.addWeighted(gX, 1, gY, 1, 0)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(10,5))
axs[0].imshow(median)
axs[0].axis('off')
axs[0].set_title('Median Filtering')
axs[1].imshow(img_sobel)
axs[1].axis('off')
axs[1].set_title('Sobel Filtering')
axs[2].imshow(img)
axs[2].axis('off')
axs[2].set_title('Original')
plt.show()
```

##### Медіанний фільтр і фільтр Собеля — це два різних типи фільтрів зображень, які використовуються в обробці зображень для різних цілей.

##### Медіанний фільтр — це нелінійний фільтр, який замінює кожен піксель зображення середнім значенням сусідніх пікселів. Медіанний фільтр зазвичай використовується для зменшення шуму та видалення дрібних артефактів на зображенні. Медіанний фільтр менш чутливий до викидів порівняно з середнім фільтром, що робить його хорошим вибором для зображень із 'солевим та перцевим' шумом . Однак медіанний фільтр також згладжує краї та дрібні деталі зображення.

##### Фільтр Sobel — це лінійний фільтр, який використовується для виявлення країв. Фільтр Собеля згортає зображення з двома окремими ядрами для обчислення градієнта інтенсивності зображення. Потім величину градієнта можна використовувати для виявлення країв на зображенні. Фільтр Sobel дуже чутливий до країв і може використовуватися для визначення країв у різних напрямках, наприклад вертикальних і горизонтальних країв.

##### Взагальному, медіанний фільтр в основному використовується для зменшення шуму та видалення дрібних артефактів, тоді як фільтр Собеля в основному використовується для виявлення країв.
