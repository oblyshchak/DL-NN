```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as clr
```

```python
img = cv2.imread('molita.jpg')
plt.imshow(img)
cv2.waitKey(0)
plt.axis('off')
plt.show
```

```python
image_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(image_RGB)
plt.show()
```

```python
image_RED = image_RGB.copy()
image_RED[:, :, (1, 2)] = 0
plt.imshow(image_RED)
plt.show()
```

```python
image_GREEN = image_RGB.copy()
image_GREEN[:, :, (0, 2)] = 0
plt.imshow(image_GREEN)
plt.show()
```

```python
image_BLUE = image_RGB.copy()
image_BLUE[:, :, (0, 1)] = 0
plt.imshow(image_BLUE)
plt.show()
```

```python
image_HSV = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
H_image = image_HSV.copy()
H_image[:, :, (1, 2)] = 0
plt.imshow(H_image)
plt.show()
```

```python
S_image = image_HSV.copy()
S_image[:, :, (0, 2)] = 0
plt.imshow(S_image)
plt.show()
```

```python
V_image = image_HSV.copy()
V_image[:, :, (0, 1)] = 0
plt.imshow(V_image)
plt.show()
```

```python
random_image = image_HSV.copy()
random_image[:, :, 1] = 0
random_image[:, :, 0] = 179
random_image_RGB = cv2.cvtColor(random_image, cv2.COLOR_HSV2RGB)
plt.imshow(random_image_RGB)
plt.show()
```

```python
random_image = image_HSV.copy()
random_image[:, :, 2] = 210
random_image[:, :, 0] = 200
random_image_RGB = cv2.cvtColor(random_image, cv2.COLOR_HSV2RGB)
plt.imshow(random_image_RGB)
plt.show()
```

```python
image_lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
```

```python
l_image = image_lab.copy()
l_image[:, :, (1, 2)] = 0
l_image_rgb = cv2.cvtColor(l_image, cv2.COLOR_LAB2BGR)
l_gray = cv2.cvtColor(l_image_rgb, cv2.COLOR_RGB2GRAY)
plt.imshow(l_gray)
plt.show()
```

```python
a_image = image_lab.copy()
a_image[:, :, 0] = 190
a_image[:, :, 2] = 20
a_image_rgb = cv2.cvtColor(l_image, cv2.COLOR_LAB2BGR)
a_gray = cv2.cvtColor(a_image_rgb, cv2.COLOR_RGB2GRAY)
cmap_a = clr.LinearSegmentedColormap.from_list('custom_blue', ['Red', 'Gray', 'Green'], N = 255)
plt.imshow(a_gray, cmap_a)
plt.show()
```

```python
b_image = image_lab.copy()
b_image[:, :, 0] = 0
b_image[:, :, 1] = 190
b_image_rgb = cv2.cvtColor(b_image, cv2.COLOR_LAB2BGR)
b_gray = cv2.cvtColor(b_image_rgb, cv2.COLOR_RGB2GRAY)
cmap_a = clr.LinearSegmentedColormap.from_list('custom_blue', ['Black', 'Yellow', 'Blue'], N = 255)
plt.imshow(a_gray, cmap_a)
plt.show()
```
